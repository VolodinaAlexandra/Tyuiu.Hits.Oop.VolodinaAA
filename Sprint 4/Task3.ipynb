<h2 style="color:DodgerBlue">Описание проекта:</h2>

----
Создать базовый класс Book в C#, который будет представлять информацию о
книгах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

Требования к базовому классу Book:
• Атрибуты: Название (Title), Автор (Author), Год издания (YearOfPublication).
• Методы:
o GetInfo(): метод для получения информации о книге в виде строки.
o Read(): метод для вывода сообщения о чтении книги.
o Borrow(): метод для вывода сообщения о выдаче книги на чтение.

Требования к производным классам:
1. Учебник (Textbook): Должен содержать дополнительные атрибуты, такие как
Предмет (Subject). Метод Read() должен быть переопределен для
добавления информации о предмете при чтении учебника.
2. Художественная литература (Fiction): Должен содержать дополнительные
атрибуты, такие как Жанр (Genre). Метод Borrow() должен быть
переопределен для добавления информации о жанре при выдаче книги на
чтение.
3. Научная литература (ScientificLiterature) (если требуется третий класс):
Должен содержать дополнительные атрибуты, такие как Область науки
(FieldOfScience). Метод GetInfo() должен быть переопределен для включения
информации об области науки в описании книги.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) создайте явную реализации интерфейса и управление зависимостями

```
public interface IBook
{
  int CurrentPage { get; }
  int TotalPages { get; }

  void TurnPageForward();
  void TurnPageBackward();
  void OpenPage(int pageNumber);
}
public interface INotReturn
{
  void NotReturn(int day);
}
public interface ITextbook
{
  string Topic {get;}
  void LearnTopic();
}

public interface IComics
{
  void LookAt();
}

public class Book: IBook
{
  public string Title;
  public string Author { get; set; }
  int _yearOfPublication; 
  public int YearOfPublication
  { 
    get
    {
      return _yearOfPublication;
    }
    set
    {
      if (value > 0)
        _yearOfPublication = value;
      else
          throw new ArgumentOutOfRangeException("Год не может быть отрицательным");
    }
  } 
  protected bool IsReading { get; set; }
  protected bool IsBorrowed { get; set; } 
  public DateTime? BorrowDate { get; set; } 
  public static int bookCount = 0;
  public int TotalPages { get; set; }
  private int _currentPage = 1;

  public int CurrentPage
  {
    get { return _currentPage; }
    private set
    {
      if (value >= 1 && value <= TotalPages)
      {
        _currentPage = value;
      }
      else
      {
        throw new ArgumentOutOfRangeException("Страница должна быть в диапазоне от 1 до " + TotalPages);
      }
    }
  }

  public Book(string _title, string _author, int _yearOfPublication,int _totalPages)
  {
    this.Title = _title;
    this.Author = _author;
    this.YearOfPublication = _yearOfPublication;
    this.IsReading = false; 
    this.IsBorrowed = false;
    this.BorrowDate = null;
    bookCount++;
    this.TotalPages = _totalPages;
  }
  public virtual string GetTitle()
  {
    return Title;
  }

  public virtual void GetInfo()
  {
    Console.WriteLine($"Название книги - {Title}, автор книги - {Author}, год публикации - {YearOfPublication},количество страниц - {TotalPages}");
  }
  public virtual void GetInfo(int shelf)
  {
    Console.WriteLine($"Книга находится в шкафу номер {shelf}");
  }

  public virtual void Read()
  {
    if (!IsReading)
    {
      Console.WriteLine($"Книга на прочтении"); 
      IsReading = true; 
    }
    else
    {
      Console.WriteLine($"Книга уже прочитана");
    }
  }

  public virtual void Borrow()
  {
    BorrowDate = DateTime.Now;
    if (!IsBorrowed)
    {
      Console.WriteLine($"Книга выдана на чтение только что");
      IsBorrowed = true; 
      
    }
    else
    {
      Console.WriteLine($"Книга уже была взята на чтение");
    }
  }

  public static int GetBookCount()
  {
    return bookCount;
  }
  
  public virtual void TurnPageForward()
  {
    CurrentPage++;
    Console.WriteLine($"Текущая страница: {CurrentPage}");
  }

  public virtual void TurnPageBackward()
  {
    CurrentPage--;
    Console.WriteLine($"Текущая страница: {CurrentPage}");
  }
  public virtual void OpenPage(int pageNumber)
  {
    CurrentPage = pageNumber;
    Console.WriteLine($"Открыта страница: {CurrentPage}");
  }

  public virtual void NotReturn(string day)
  {
    Console.WriteLine($"Книга задержана на {day}");
  }
}

public class Textbook : Book,IBook,ITextbook
{
    public string Topic {get;set;}
    public string Subject { get; set; }

    public Textbook(string _title, string _author, int _yearOfPublication,int _totalPages, string _subject,string _topic) : base(_title, _author, _yearOfPublication, _totalPages)
    {
        Subject = _subject;
        Topic = _topic;
    }

    public override void Read()
    {
        if (IsReading)
        {
          Console.WriteLine($"Книга {Title} по предмету {Subject} уже была прочитана");
        }
        else
        {
          Console.WriteLine($"Книга {Title} по предмету {Subject} сейчас читается");
          IsReading = true;
        }
    }
    public override void TurnPageForward()
    {
      base.TurnPageForward();
      Console.WriteLine($"Изучаю материал на странице {CurrentPage} учебника."); 
    }

    public override void TurnPageBackward()
    {
      base.TurnPageBackward();
      Console.WriteLine($"Возвращаюсь к материалу на странице {CurrentPage} учебника."); 
    }

    public void LearnTopic()
    {
      Console.WriteLine($"Изучаю тему '{Topic}' в учебнике '{Title}'");
    }
}

public class Fiction : Book,IBook,INotReturn
{
  public string Genre { get; set; }

  public Fiction(string _title, string _author, int _yearOfPublication,int _totalPages, string _genre) : base(_title, _author, _yearOfPublication,_totalPages)
  {
    Genre = _genre;
  }

  public override void Borrow()
  {
    BorrowDate = DateTime.Now;
    if (IsBorrowed)
    {
      Console.WriteLine($"Книга {Title} в жанре {Genre} уже взята на чтение {BorrowDate}");
    }
    else
    {
      Console.WriteLine($"Книга {Title} в жанре {Genre} была взята на чтение {BorrowDate}");
      IsBorrowed = true;
    }
  }
  public override void TurnPageForward()
  {
    base.TurnPageForward();
    Console.WriteLine($"Погружаюсь в мир книги на странице {CurrentPage}.");
  }

  public override void TurnPageBackward()
  {
    base.TurnPageBackward();
    Console.WriteLine($"Перечитываю интересный момент на странице {CurrentPage}.");
  }
  public override void NotReturn(string day)
  {
    Console.WriteLine($"Книга задержана на {day}");
  }
  public void NotReturn(int day)
  {
    Console.WriteLine($"Задержка книги: {day} дней");
  }
}

public class ScientificLiterature : Book,IBook
{
  private string FieldOfSciense { get; set; }
  private List<Book> sources = new List<Book>();


  public ScientificLiterature(string _title, string _author, int _yearOfPublication,int _totalPages, string _fieldofscience) : base(_title, _author, _yearOfPublication,_totalPages)
  {
    FieldOfSciense = _fieldofscience;
  }

  public void AddSource(Book book)
  {
    this.sources.Add(book);
  }

  public override void GetInfo()
  {
    base.GetInfo();
    Console.WriteLine($"Область науки - {FieldOfSciense}");
  }
  
  public void PrintSources()
  {
    string s = "";
    for(int i = 0; i < sources.Count; i++){
      s += sources[i].GetTitle();
      s += ", ";
    }
    Console.WriteLine($"Книга {this.GetTitle()} ссылается на: {s}");
  }
  public override void TurnPageForward()
  {
    base.TurnPageForward();
    Console.WriteLine($"Изучаю интересную статью на странице {CurrentPage}.");
  }

  public override void TurnPageBackward()
  {
    base.TurnPageBackward();
    Console.WriteLine($"Перечитываю важный для понимания темы момент на странице {CurrentPage}.");
  }
}
public class Comics : Fiction,IComics
{
    public string Hero {get;set;}
    public Comics(string _title, string _author, int _yearOfPublication, string _genre,int _totalPages,string _hero) : base(_title, _author, _yearOfPublication,_totalPages,_genre)
    {
        this.Hero = _hero;
    }
    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Этот комикс содержит героя: {Hero}");
    }
    public override void TurnPageForward()
    {
      base.TurnPageForward();
      Console.WriteLine($"Погружаюсь в мир книги на странице {CurrentPage}.");
    }

    public override void TurnPageBackward()
    {
      base.TurnPageBackward();
      Console.WriteLine($"Перечитываю интересный момент на странице {CurrentPage}.");
    }
    void IComics.LookAt()
    {
      Console.WriteLine($"Смотрю в комиксе {Title} на героя {Hero} на странице {CurrentPage}");
    }
}
public class BookCollection<T> where T : Book
{
    private List<T> books = new List<T>();

    public void AddBook(T book)
    {
        books.Add(book);
    }

    public void PrintAllBooks()
    {
        foreach (T book in books)
        {
            Console.WriteLine($"'{book.GetTitle()}' находится в библиотеке");
        }
    }
}
public class Library
{
  private readonly IBook _book;
  public Library(IBook book)
  {
    _book = book;
  }
  public void UseBook()
  {
    _book.OpenPage(25);
    _book.TurnPageForward();
    Console.WriteLine($"В этой книге {_book.TotalPages} страниц");
  }
}
Book myBook = new Book("Гарри Поттер", "Джоан Роулинг", 1999,489);
myBook.GetInfo();
myBook.GetInfo(5);
myBook.Read();
myBook.Borrow();
Console.WriteLine("\n");
Textbook math = new Textbook("Математика 6 класс","Ященко", 2024, 268,"Математика","Пропорции");
math.GetInfo();
math.Read();
math.Borrow();
math.OpenPage(28);
math.TurnPageForward();
Console.WriteLine("\n");
Fiction dracula = new Fiction("Дракула","Брэм Стокер",1897,502,"Ужасы");
dracula.GetInfo();
dracula.Read();
dracula.Borrow();
dracula.NotReturn("тридцать дней");
dracula.NotReturn(30);
Console.WriteLine("\n");
ScientificLiterature physics = new ScientificLiterature("Теория всего","Стивен Хокинг",2002,216,"Физика");
physics.AddSource(math);
physics.AddSource(myBook);
physics.AddSource(dracula);
physics.PrintSources();
physics.GetInfo();
physics.Read();
physics.Borrow();
Console.WriteLine("\n");
Comics batman = new Comics("Batman","Боб Кейн",1939,"комикс",150,"бетмен");
batman.GetInfo();
batman.Borrow();
batman.Read();
IComics superman = new Comics( "Superman", "Джерри Сигел", 1938,"Комикс", 150, "superman");
superman.LookAt();
Console.WriteLine("\n");
Console.WriteLine($"Всего книг в библиотеке:{Book.GetBookCount()}");
BookCollection<Book> bookCollection = new BookCollection<Book>();
bookCollection.AddBook(math);
bookCollection.AddBook(dracula);
bookCollection.AddBook(physics);
bookCollection.AddBook(batman);
Console.WriteLine("\n");
IBook myDiary = new Book("Село Степанчиково","Достоевский",1859,560);
Library myDiaryLibrary = new Library(myDiary);
myDiaryLibrary.UseBook();
Console.WriteLine("Все книги в библиотеке:");
bookCollection.PrintAllBooks();
```
